```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IHME_GBD_2019_SMOKING_TOB_1990_2019_NUM_SMOKERS_Y2021M05D27.CSV')
df
```

```python
df.dropna()
```

```python
series1 = df['val']
series1
```

```python
# obliczanie Å›redniej

np.mean(series1)
```

```python
#obliczanie mediany 

np.median(series1)
```

```python
#Odchylenie standardowe

np.std(series1)
```

```python
#Wariancja 

np.var(series1)
```

```python
series2 = df['upper']

series2
```

```python
#Koleracja

correlation = np.corrcoef(series1,series2) [0,1]
print(correlation)
```

```python
#Kowariancja 

covariance = np.cov(series1,series2) [0,1]
covariance
```
